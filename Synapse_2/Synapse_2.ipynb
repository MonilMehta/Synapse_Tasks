```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv")
```

```python
df.dropna(inplace = True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df["country"].value_counts().head(5)
```

```python
top_countries = ['United States', 'India', 'United Kingdom', 'Canada', 'Spain']

top_country_df = df[df['country'].isin(top_countries)]
top_country_df
```

```python
pie_df=top_country_df['country'].value_counts()
```

```python
pie_df
```

```python
pie_df.to_csv("Pie_chart.csv")
```

```python
pie_df.plot(kind='pie', labels=pie_df.index, autopct='%1.1f%%')
plt.title("Country's Pie Chart")
plt.axis('equal')
plt.show()
```
